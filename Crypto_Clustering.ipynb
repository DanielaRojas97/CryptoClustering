```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

#Select numerical columns to scale (replace 'column1', 'column2', etc. with your actual column names)
numerical_cols = df_market_data.select_dtypes(include=['float64', 'int64']).columns
# Initialize the StandardScaler
scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(df_market_data[numerical_cols])

# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaled_data, columns=numerical_cols, index=df_market_data.index)
# Display the first few rows of the scaled DataFrame
df_scaled.head()


# # Copy the crypto names from the original DataFrame (my code already does it)
# Set the coin_id column as index

# Display the scaled DataFrame
df_scaled.index
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 11)  # You can adjust the range as needed
```

```python
# Create an empty list to store the inertia values
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list


# Create a list to store inertia values and the range of k values
inertia = []

# Loop through the range of k values
for k in k_values:
    # Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fit the model to the data using `df_market_data_scaled`
    model.fit(df_scaled)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

#zip: pairs each ( k ) value with its corresponding inertia value, creating tuples of the form (k, inertia).
#dict: This converts the zipped tuples into a dictionary.
elbow_data_dict = dict(zip(k_values, inertia))

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(list(elbow_data_dict.items()), columns=['k', 'inertia'])
elbow_data_df

```

```python
# Plot a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_data_df.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Method for Optimal k',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    width=800,
    height=400
)

# Display the plot
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 3

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
#Set the optimal number of clusters
optimal_k = 3  

# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=optimal_k, random_state=42) 
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_scaled)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)  # Reduce to 3 components for visualization
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame down to three principal components.
pca_result = pca.fit_transform(df_scaled[numerical_cols])  # Use only numerical columns for PCA

# Create a DataFrame for PCA results
df_pca = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2', 'PCA3'], index=df_scaled.index)  # Update to include PCA3

# Add the cluster labels to the PCA DataFrame
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# Create a DataFrame to display the explained variance
explained_variance_df = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(explained_variance))],
    'Explained Variance Ratio': explained_variance
})

# Display the explained variance DataFrame
explained_variance_df
```

```python
# Total explained variance for the selected principal components
total_explained_variance = explained_variance.sum()

# Display the total explained variance
total_explained_variance
```

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 11)
```

```python
# Create an empty list to store the inertia values
inertia = [] #Create a list to store inertia values

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

#Loop through the range of k values
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the scaled PCA data
    model.fit(df_pca[['PCA1', 'PCA2', 'PCA3']])  #2. Fit the model to the data using `df_market_data_pca`
    
    # Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_dict = dict(zip(k_values, inertia))

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(list(elbow_data_dict.items()), columns=['k', 'inertia'])
elbow_data_df
```

```python
# Plot a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_data_df.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Method for Optimal k',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    width=800,
    height=400
)

# Display the plot
elbow_plot
```
